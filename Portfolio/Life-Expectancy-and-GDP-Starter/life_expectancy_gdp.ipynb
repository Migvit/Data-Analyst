```python
import pandas as pd

gpd = pd.read_csv('all_data.csv')
print(gpd.head())
```

```python
# Display basic info
print(gpd.info())

# Compute descriptive statistics
desc_stats = gpd.describe()
print(desc_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram for Life Expectancy
sns.histplot(gpd["Life expectancy at birth (years)"], bins=10, kde=True, ax=axes[0])
axes[0].set_title("Distribution of Life Expectancy")
axes[0].set_xlabel("Life Expectancy (Years)")
axes[0].set_ylabel("Frequency")

# Histogram for GDP
sns.histplot(gpd["GDP"], bins=10, kde=True, ax=axes[1])
axes[1].set_title("Distribution of GDP")
axes[1].set_xlabel("GDP (in USD)")
axes[1].set_ylabel("Frequency")

# Box Plot for Life Expectancy and GDP
sns.boxplot(data=gpd[["Life expectancy at birth (years)", "GDP"]], ax=axes[2])
axes[2].set_title("Box Plot of Life Expectancy & GDP")

# Adjust layout and display plots
plt.tight_layout()
plt.show()
```

```python

# --- 2. VISUALIZING TRENDS ---
plt.figure(figsize=(12, 6))
sns.lineplot(data=gpd, x="Year", y="Life expectancy at birth (years)", hue="Country", marker="o")
plt.title("Life Expectancy Over Time")
plt.ylabel("Life Expectancy (Years)")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=gpd, x="Year", y="GDP", hue="Country", marker="o")
plt.title("GDP Over Time")
plt.ylabel("GDP (USD)")
plt.yscale("log")  # Log scale to handle large GDP differences
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# --- 3. CORRELATION ANALYSIS ---
correlation = gpd[["Life expectancy at birth (years)", "GDP"]].corr()
print(correlation)

sns.scatterplot(data=gpd, x="GDP", y="Life expectancy at birth (years)", hue="Country")
plt.xscale("log")  # GDP has large variations, log scale helps visualization
plt.title("GDP vs. Life Expectancy")
plt.show()

# --- 4. OUTLIER DETECTION ---
plt.figure(figsize=(8, 5))
sns.boxplot(data=gpd, x="Life expectancy at birth (years)")
plt.title("Box Plot for Life Expectancy")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=gpd, x="GDP")
plt.title("Box Plot for GDP")
plt.xscale("log")  # Log scale for better visibility
plt.show()
```
