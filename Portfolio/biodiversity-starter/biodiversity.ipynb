```python
import pandas as pd

# Caminhos dos arquivos
file_observations = "observations.csv"
file_species = "species_info.csv"

# Carregar os CSVs
df_observations = pd.read_csv(file_observations)
df_species = pd.read_csv(file_species)

# Ver as primeiras linhas
print(df_observations.head())
print(df_species.head())

# Verificar informações gerais dos DataFrames
print(df_observations.info())
print(df_species.info())

# Checar valores nulos
print(df_observations.isnull().sum())
print(df_species.isnull().sum())

```

```python
df_merged = pd.merge(df_observations, df_species, on="scientific_name", how="left")
df_merged["conservation_status"] = df_merged["conservation_status"].fillna("Não Avaliado")

print(df_merged.info())
print(df_merged.isnull().sum())
df_merged.to_csv("dados_limpos.csv", index=False)
```

```python
df_merged.head()
```

```python
observacoes_por_parque = df_merged.groupby("park_name")["observations"].sum().sort_values(ascending=False)
print(observacoes_por_parque)
import matplotlib.pyplot as plt

observacoes_por_parque.plot(kind="bar", figsize=(10,5), color="skyblue")
plt.title("Número total de observações por parque")
plt.xlabel("Parque")
plt.ylabel("Número de Observações")
plt.xticks(rotation=45)
plt.show()
```

```python
especies_mais_observadas = df_merged.groupby("scientific_name")["observations"].sum().sort_values(ascending=False).head(10)
print(especies_mais_observadas)
especies_mais_observadas.plot(kind="barh", figsize=(10,5), color="green")
plt.title("Top 10 Espécies Mais Observadas")
plt.xlabel("Número de Observações")
plt.ylabel("Espécie")
plt.gca().invert_yaxis()
plt.show()
```

```python
categorias_contagem = df_merged["category"].value_counts()
print(categorias_contagem)
categorias_contagem.plot(kind="pie", autopct="%1.1f%%", figsize=(8,8), colormap="viridis")
plt.title("Distribuição das Categorias de Espécies")
plt.ylabel("")
plt.show()
```

```python
status_conservacao = df_merged["conservation_status"].value_counts()
print(status_conservacao)
status_conservacao.plot(kind="bar", color="red", figsize=(8,5))
plt.title("Status de Conservação das Espécies")
plt.xlabel("Status de Conservação")
plt.ylabel("Número de Espécies")
plt.xticks(rotation=45)
plt.show()
```

```python
df_merged["ameacada"] = df_merged["conservation_status"] != "Não Avaliado"
observacoes_ameacadas = df_merged.groupby("ameacada")["observations"].sum()
print(observacoes_ameacadas)
observacoes_ameacadas.plot(kind="bar", color=["blue", "red"])
plt.title("Observações de Espécies Ameaçadas vs. Não Ameaçadas")
plt.xlabel("Ameaçada?")
plt.ylabel("Número de Observações")
plt.xticks(ticks=[0,1], labels=["Não", "Sim"], rotation=0)
plt.show()
```
