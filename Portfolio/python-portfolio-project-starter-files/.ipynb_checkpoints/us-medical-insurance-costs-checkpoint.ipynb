# U.S. Medical Insurance Costs

```python
import csv

# Variables
ages = []
sexs = []
bmis = []
childrens = []
smokers = []
regions = []
charges = []

# Import CSV
with open ("insurance.csv") as insurance:
    reader = csv.reader(insurance)
    #for row in reader: 
        #print(row)

# Set Variables
with open("insurance.csv") as insurance:
    reader = csv.DictReader(insurance)

    for row in reader:
        # Adicionar os valores às listas correspondentes, convertendo quando necessário
        ages.append(int(row['age']))  # Converter para int
        sexs.append(row['sex'])  # String
        bmis.append(float(row['bmi']))  # Converter para float
        childrens.append(int(row['children']))  # Converter para int
        smokers.append(row['smoker'])  # String
        regions.append(row['region'])  # String
        charges.append(float(row['charges']))  # Converter para float

# Exemplo: Verificar as primeiras entradas de cada lista
"""print("Idades:", ages[:5])
print("Sexos:", sexs[:5])
print("BMIs:", bmis[:5])
print("Filhos:", childrens[:5])
print("Fumantes:", smokers[:5])
print("Regiões:", regions[:5])
print("Custos:", charges[:5])"""

# Analysis

# Average BMI x Charges

def bmixcharges(bmi, charges):
    bmis_avg = sum(bmi) / len(bmi)  
    charges_avg = sum(charges) / len(charges)
    
    return bmis_avg, charges_avg


avg_bmi, avg_charges = bmixcharges(bmis, charges)
print("Average BMI: " + str(avg_bmi))
print("Average Charges:" + str(avg_charges))


# Age per Region
def agexregion(ages, regions):
    region_age_map = {}
    region_avg_map = {}

    for age, region in zip(ages, regions):
        if region not in region_age_map:
            region_age_map[region] = []  
        region_age_map[region].append(age)
    for region, age_list in region_age_map.items():
        region_avg_map[region] = sum(age_list) / len(age_list)  

    return region_avg_map

region_averages = agexregion(ages, regions)

# Exibir os resultados
print("Average age by region:")
for region, avg in region_averages.items():
   print(f"{region}: {avg:.2f} yo")    


# Função para calcular a porcentagem de sexo masculino e feminino por faixas etárias de 5 anos
def gender_percentage_by_age_range(ages, sexes):
    age_sex_map = {}  # Dicionário para agrupar sexos por faixa etária

    # Agrupar sexos por faixa etária
    for age, sex in zip(ages, sexes):
        age_range = f"{(age // 5) * 5}-{(age // 5) * 5 + 4}"  # Ex.: 0-4, 5-9, etc.
        if age_range not in age_sex_map:
            age_sex_map[age_range] = {"male": 0, "female": 0}  # Inicializar contadores
        if sex == "male":
            age_sex_map[age_range]["male"] += 1
        elif sex == "female":
            age_sex_map[age_range]["female"] += 1

    # Calcular as porcentagens para cada faixa etária
    age_sex_percentage = {}
    for age_range, counts in age_sex_map.items():
        total = counts["male"] + counts["female"]
        if total > 0:  # Evitar divisão por zero
            male_percentage = (counts["male"] / total) * 100
            female_percentage = (counts["female"] / total) * 100
        else:
            male_percentage = female_percentage = 0
        age_sex_percentage[age_range] = {
            "male_percentage": male_percentage,
            "female_percentage": female_percentage,
        }

    return age_sex_percentage


# Calcular a porcentagem de sexo por faixa etária
gender_percentages = gender_percentage_by_age_range(ages, sexs)

# Exibir os resultados
print("Percentage of sex by age groups (5 years):")
for age_range, percentages in gender_percentages.items():
    print(
        f"Age group {age_range}: {percentages['male_percentage']:.2f}% masculine, "
        f"{percentages['female_percentage']:.2f}% feminine"
    )
```
